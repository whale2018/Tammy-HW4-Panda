### Heroes Of Pymoli Data Analysis
* Of the 1163 active players, the vast majority are male (84%). There also exists, a smaller, but notable proportion of female players (14%).

* Our peak age demographic falls between 20-24 (44.8%) with secondary groups falling between 15-19 (18.60%) and 25-29 (13.4%).

```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# Raw data file
file_to_load = "Resources/purchase_data.csv"

# Read purchasing file and store into pandas data frame
purchase_data = pd.read_csv(file_to_load)
purchase_data.head()
```

## Player Count

* Display the total number of players


```python
total_players = len(purchase_data["SN"].unique())
table = pd.DataFrame([{"Total Players": total_players}])
table
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
pd.options.display.float_format = '${:,.2f}'.format
unique = len(purchase_data["Item ID"].unique())
ave_price = purchase_data["Price"].mean()
no_purchase = purchase_data["Purchase ID"].count()
total_rev = purchase_data["Price"].sum()
summary_table = pd.DataFrame([{"Number of Unique Items": unique, 
                             "Average Price": ave_price,
                             "Number of Purchases": no_purchase,
                              "Total Revenue": total_rev}])
organized_summary_table = summary_table[["Number of Unique Items", "Average Price", "Number of Purchases", "Total Revenue"]]
organized_summary_table
```

## Gender Demographics

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
remove_dup_data = purchase_data.drop_duplicates(subset= 'SN', keep='first', inplace=False)

combined = remove_dup_data.groupby(["Gender"])

total_players_nodup = remove_dup_data["SN"].count()

gender_count = remove_dup_data["Gender"].value_counts()

percentage = ((gender_count / total_players_nodup)*100).map("{:.2f}%".format)

summary_table_gender = pd.DataFrame({"Percentage of Players": percentage, 
                                     "Total Count": gender_count})
summary_table_gender
                               
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, etc. by gender


* For normalized purchasing, divide total purchase value by purchase count, by gender


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
combined = purchase_data.groupby(["Gender"])

purchase_count_gender = combined["Purchase ID"].count()

ave_purchase_price = combined["Price"].mean().map("${:.2f}".format)

total_purchase_value = combined["Price"].sum().map("${:,.2f}".format)

normalized_purchasing = (combined["Price"].sum() / combined["Purchase ID"].count()).map("${:.2f}".format)

summary_table_purchasing = pd.DataFrame({"Purchase Count": purchase_count_gender, 
                                         "Average Purchase Price": ave_purchase_price,
                                         "Total Purchase Value": total_purchase_value,
                                         "Normalized Totals": normalized_purchasing})
summary_table_purchasing
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
# Establish bins for ages
age_bins = [0, 9.90, 14.90, 19.90, 24.90, 29.90, 34.90, 39.90, 99999]
group_names = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

remove_dup_data["Age Group"] = pd.cut(remove_dup_data["Age"], age_bins, labels=group_names)

combined_count_age = remove_dup_data.groupby(["Age Group"])

agegroup_count = combined_count_age["Purchase ID"].count()

percen_players = ((agegroup_count / total_players_nodup)*100).map("{:.2f}%".format)

summary_table_age = pd.DataFrame({"Percentage of Players": percen_players, 
                                  "Total Count": agegroup_count})
summary_table_age                 
    
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, etc. in the table below


* Calculate Normalized Purchasing


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
age_bins = [0, 9.90, 14.90, 19.90, 24.90, 29.90, 34.90, 39.90, 99999]
group_names = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

purchase_data["Age Group"] = pd.cut(purchase_data["Age"], age_bins, labels=group_names)

combined_count_agegroup = purchase_data.groupby(["Age Group"])
combined_count_agegroup.head()

purchase_count_age = combined_count_agegroup["Purchase ID"].count()
purchase_count_age

ave_purchase_price_age = combined_count_agegroup["Price"].mean().map("${:.2f}".format)
ave_purchase_price_age

total_purchase_value_age = combined_count_agegroup["Price"].sum().map("${:,.2f}".format)
total_purchase_value_age

normalized_purchasing_age = (combined_count_agegroup["Price"].sum() / combined_count_agegroup["Purchase ID"].count()).map("${:.2f}".format)
normalized_purchasing_age

summary_table_age = pd.DataFrame({"Purchase Count": purchase_count_age, 
                                  "Average Purchase Price": ave_purchase_price_age,
                                  "Total Purchase Value": total_purchase_value_age,
                                  "Normalized Totals": normalized_purchasing_age})
summary_table_age

```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
combined_spenders = purchase_data.groupby(["SN"])

purchase_count_spenders = combined_spenders["Purchase ID"].count()


average_purchase_price_spenders = combined_spenders["Price"].mean().map("${:.2f}".format)

total_purchase_value_spenders = combined_spenders["Price"].sum().map("${:,.2f}".format)


summary_table_spenders = pd.DataFrame({"Purchase Count": purchase_count_spenders,
                                       "Average Purchase Price": average_purchase_price_spenders,
                                       "Total Purchase Value": total_purchase_value_spenders})

Total_value_decending = summary_table_spenders.sort_values("Purchase Count", ascending=False)
Total_value_decending.head()
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
purchase_type = purchase_data[["Item ID", "Item Name", "Price"]]

purchase_group = purchase_type.groupby(["Item ID", "Item Name"])

purchase_count_popular = purchase_group["Item ID"].count()

ip = purchase_type["Price"].mean()

total_purchase_value_popular = purchase_group["Price"].sum()

pd.options.display.float_format = '${:,.2f}'.format

summary_table_popular = pd.DataFrame({"Purchase Count": purchase_count_popular, 
                                      "Item Price": ip,
                                      "Total Purchase Value": total_purchase_value_popular.map("${:.2f}".format)})

popular_table_descending = summary_table_popular.sort_values("Purchase Count", ascending=False)
popular_table_descending.head()
```

## Most Profitable Items

* Sort the above table by total purchase value in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the data frame



```python
purchase_type = purchase_data[["Item ID", "Item Name", "Price"]]

purchase_group = purchase_type.groupby(["Item ID", "Item Name"])

purchase_count_popular = purchase_group["Item ID"].count()

ip = purchase_type["Price"].mean()

total_purchase_value_popular = purchase_group["Price"].sum()

summary_table_popular = pd.DataFrame({"Purchase Count": purchase_count_popular, 
                                      "Item Price": ip,
                                      "Total Purchase Value": total_purchase_value_popular})

popular_table_descending = summary_table_popular.sort_values("Total Purchase Value", ascending=False)
popular_table_descending.head()
```
